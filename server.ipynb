```python
#@title Client Code

import os
import shutil
from gradio_client import Client


def generate_and_save_image(
    positive_prompt,
    negative_prompt="",
    width=1024,
    height=1024,
    steps=9,
    save_dir="z_image_turbo_images"
):
    global client
    # Call API
    result = client.predict(
        positive_prompt=positive_prompt,
        negative_prompt=negative_prompt,
        width=width,
        height=height,
        steps=steps,
        api_name="/generate_image"
    )

    # Ensure save folder exists
    os.makedirs(save_dir, exist_ok=True)

    # Extract filename from returned temp path
    filename = os.path.basename(result)

    # Final save path
    final_path = os.path.join(save_dir, filename)

    # Copy file locally
    shutil.copy(result, final_path)

    return final_path
#copy paste the public  url
gradio_url="https://54a0319f1d589fsdf128ff.gradio.live/"
client = Client(gradio_url)

img_path = generate_and_save_image(
    "a cute girl with short hair in a flower field"
)

print("Saved to:", img_path)
```
